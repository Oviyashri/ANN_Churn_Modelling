```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path='../input/churn-modellingcsv/Churn_Modelling.csv'
df=pd.read_csv(path,sep=",")
```

```python
df.head()
```

```python
X=df.iloc[:,3:13]
y=df.iloc[:,13]
```

```python
X
```

```python
y
```

```python
# Create dummy variables
geography=pd.get_dummies(X["Geography"],drop_first=True)
gender=pd.get_dummies(X["Gender"],drop_first=True)
```

```python
geography
```

```python
gender
```

```python
# Concatenate Dataframes
X=pd.concat([X,geography,gender],axis=1)
X
```

```python
# Drop unnecessary columns
X=X.drop(['Geography','Gender'],axis=1)
X
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#ANN
#Import keras libraries
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
# Initialising the ANN
Classifier=Sequential()
```

```python
## Adding the I/P layer & first hidden layer
Classifier.add(Dense(units=6, kernel_initializer='he_uniform', activation='relu', input_dim=11))
```

```python
## Second hidden layer
Classifier.add(Dense(units=6, kernel_initializer='he_uniform', activation='relu'))
```

```python
# Adding the O/P layer
Classifier.add(Dense(units=1,kernel_initializer='glorot_uniform', activation='sigmoid'))
```

```python
#Compiling ANN
Classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Fitting ANN to training dataset
model=Classifier.fit(X_train,y_train,validation_split=0.33, batch_size=10, epochs=20)
```

```python
# Predicting the Test set
y_pred=Classifier.predict(X_test)
y_pred=(y_pred>0.5)
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
score
```
